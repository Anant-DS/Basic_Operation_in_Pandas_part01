```python
# By using sorting method we can get sorted table 
# But when we call dataframe again we get old table
# So for that we have to use 'inplace' to change table
```

```python
df.sort_index(axis=1,inplace=True)
df
```

```python
# we can short vales also
df.sort_values('SepalLengthCm')
```

```python
# And suppose we call dataframe again it will show us changes DF 
# So we have to use 'inplace = true'

df.reset_index(inplace=True)
```
